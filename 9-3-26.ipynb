# **Exception handling**

```python
# it is used to handle run time errors
```

```python
try:
  a=10
  b=0
  print(a/b)
except:
  print("not divided by zero")
```

```python
try:
  a=10
  b=0
  print(a/b)
except:
  print("not divided by zero")
print("hello")
```

```python
try:
  a=10
  b=0
  print(a/b)
except ZeroDivisionError as z:
  print(z)
```

```python
# type error
try:
  a=10
  b="hi"
  print(a+b)
except:
  print("type error")
```

```python
# value error
try:
  a=int(input("enter 1 number"))
  b=int(input("enter 2 number"))
  print("o/p: ",a+b)
except ValueError as v:
  print(v)
```

```python
# index error
try:
  l=[10,20,30]
  print(l[4])
except IndexError as i:
  print(i)
```

```python
# keyerror
try:
  student={
      "name":"chiku",
      "roll":658,
      "branch":"CSE"
  }
  print(student["city"])
except KeyError as k:
  print(k)
```

```python
# name error
try:
  print(x)
except NameError as n:
  print(n)
```

```python
# filenotfound error
try:
  x=open("data.txt")
  print(x)
except FileNotFoundError as f:
  print(f)
```

```python
try:
  a=int(input("enter "))
  b=input("enter ")
  print(a/b)
except ZeroDivisionError as z:
  print("zero error ",z)
except ValueError as v:
  print("value error ",v)
except TypeError as t:
  print("type error ",t)
# except (ZeroDivisionError,ValueError) as f:
#   print(f)
```

```python
# else error
try:
  a=int(input("enter 1 number"))
  b=int(input("enter 2 number"))
  print("o/p: ",a+b)
except ValueError as v:
  print(v)
else: #it only execute when exception is not there
  print("program execute")
finally:  #it executes when there is exception and also in not exception case in program
  print("final block ")
```

```python
# generic exception
try:
  a=10
  b="HI"
  print(a/b)
except Exception as e:
  print(e)
```
