# ***access specifier***

```python
# inside the class
class student:
  def __init__(self,name,roll,sec):
    self.name=name  #public
    self._roll=roll #protected
    self.__sec=sec  #private
  def show(self):
    print(f"{self.name} | {self._roll} | {self.__sec}")

  def _show_roll(self):
    print(self._roll)

  def __show_section(self):
    print(self.__sec)

s=student("chiku",658,"j")
s.show()
s._show_roll()
# s.__show_section()

# print(s.name)
# print(s._roll)
# print(s.__sec)
```

```python
# derived class
class info(student):
  def display(self):
    # print(f"{self.name} | {self._roll} | {self.__sec}") # because of private __sec
    print(f"{self.name} | {self._roll}")

i=info("hari",655,"j")
i.display()
i.show()
i._show_roll()
# i.__show_section()
```

```python
class bank:
  def __init__(self,account_holder_name,branch,amount):
    self.name=account_holder_name
    self._branch=branch
    self.__amount=amount
  def show(self):
    print(self.name)
    print(self._branch)
    print(self.__amount)
b=bank("chiku","SBI",10000)
b.show()
print(b._bank__amount)
```

```python
class smallBank(bank):
  def display(self):
    print(self.name)
    print(self._branch)
    print(self._bank__amount) # objectname._classname__variablename(syntax)
s=smallBank("Ayush","BOB",20000)
s.display()
print("-------------------------------------")
s.show()
```

```python
class employee:
  def __init__(self,name,dept,sal):
    self.emp_name=name
    self._emp_dept=dept
    self.__emp_sal=sal
  def show(self):
    print(self.emp_name)
    print(self._emp_dept)
  def get_access(self):
    print("employee salary is : ",self.__emp_sal)
  def set_access(self,sal):
    self.__emp_sal=sal
    print("eployee salary after set : ",self.__emp_sal)
e=employee("ayush","SBI",20000)
e.show()
e.get_access()
e.set_access(3000)
```

```python
class update(employee):
  def update_salary(self,sal):
    self.__emp_sal=sal
    print("updated salary is : ",self.__emp_sal)

u=update("chiku","BOB",50000)
u.update_salary(30000)
u.show()
u.get_access()
```
