```python
def stu(info):
  print("name is ",info["name"])
x={"name":"chiku"}
stu(x)
```

```python
# arbitary positional arguments
def stu(*name):
  for i in name:
    print("good morning",i)
stu("atul","amit","shreya")
```

```python
# keyword arguments
def stu(info):
  print("name of student",info["name"])
  print("roll no of student",info["roll"])
x={"name":"ram","roll":23}
stu(x)
```

```python
# arbitary keyword arguments
def stu(**info):
  print("name of student",info["name"])
  print("roll no of student",info["roll"])
stu(name="ram",roll=23)
```

```python
"""
practice questions-1
customer name=positional
discount=default
any number of item prices-arbitary positional

print-
total-
discount amount-
final amount-
"""
```

```python
def shop(name,*item_prices,discount=10):
  total=sum(item_prices)
  print(f"your name is: {name}")
  print(f"your total is: {total}")
  print(f"your discount is: {(total*discount)/100}")
  print(f"your final amount is: {total-discount}")

shop("chiku",20,30,40)
```

```python
"""
practice question-2
bank transcation

balance=10,000(global)
deposite=local
withdraw=local
"""
```

```python
balance=10000
def deposite(amount):
  global balance
  print(f"you deposited 💸{amount} in your bank account")
  balance +=amount
  print(f"your curent balance is: 💸{balance}\n")
def withdraw(amount):
  global balance
  print(f"you withdraw 💸{amount} in your bank account")
  balance -=amount
  print(f"your current balance is: 💸{balance}\n")
deposite(1000)
withdraw(2000)
deposite(5000)
```

# **Lambda Function**

```python
# syntax-lambda arguments: expression
"""
name less function
small function
easy way for performing calculation
"""
add=lambda a,b : a+b
print(add(10,20))
sub=lambda c,d : c-d
print(sub(2,3))
square=lambda a : a*a # or a**2
print(square(15))
power=lambda a : a**3
print(power(3))
```

# **map**

```python
"""
map:-
every element it performing some operations
syntax-
lists=[1,2,3,4,5]
square=map(lambda x : x*x , lists)  where, lists=iterable(list,tuple,dictionary,set) lambda=function
print(list(square))

filter:-
it only returns true value
"""
```

```python
lists=[1,2,3,4,5]
square=map(lambda x : x*x , lists)
print(list(square))
```

```python
list2=["2","4","5"]
convert_number=map(lambda x : int(x), list2)
print(list(convert_number))
print(type(list(convert_number)))
```

```python
list3=[1,2,3,4,5,6]
result=filter(lambda x : x%2 == 0, list3)
print(list(result))
```

```python
list4=["ram","hari","hi","chiku"]
name_length3=filter(lambda x: len(x)>3 ,list4)
print(list(name_length3))
```
