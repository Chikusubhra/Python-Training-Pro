```python
create database gietu;
use gietu;
create table student(
rollno int,
stu_name varchar(20),
email varchar(30)
);
show tables;
desc student;
insert into student(rollno,stu_name,email) values(1,"chiku","chiku@gmail.com");
insert into student(rollno,stu_name,email) values(2,"piku","piku@gmail.com");

select * from student;
alter table student add city varchar(20);
desc student;
alter table student modify stu_name varchar(25);
desc student;
-- update student set stu_name="daku", email="daku@gmail.com" where rollno=2;
delete from student where rollno=1;
insert into student(rollno,stu_name,email) values(3,"aryan","aryan@gmail.com");
insert into student(rollno,stu_name,email) values(4,"omm","omm@gmail.com");
insert into student(rollno,stu_name,email) values(5,"ram","ram@gmail.com");
select * from student;

SET SQL_SAFE_UPDATES = 0;
update student set stu_name="daku", email="daku@gmail.com" where rollno=2;

select * from student;

-- truncate table student;
alter table student add mark varchar(20);
select * from student;
update student set city="pune" where rollno=1;
update student set city="thane" where rollno=2;
update student set city="panvel" where rollno=3;
update student set city="navi-mumbai" where rollno=4;

select * from student where city="thane";

select * from student order by city desc;
select city, count(*) from student group by city;


select stu_name, avg(mark) from student group by stu_name;

alter table student add course varchar(20);
desc student;


update student set course="java" where rollno=2;
update student set course ="python" where rollno=4;

update student set mark=12 where rollno=1;
update student set mark=13 where rollno=2;
update student set mark=11 where rollno=3;
update student set mark=15 where rollno=4;


select stu_name,min(mark) from student group by stu_name;

drop database gietu


-- drop table student

-- truncate table student


```
