```python
class number:
  def __iter__(self):
    self.num=1
    return self
  def __next__(self):
    if self.num<=5:
      x=self.num
      self.num += 1
      return x
    else:
      raise StopIteration
n=number()
it=iter(n)
while True:
  try:
    print(next(it))
  except StopIteration as s:
    print(s)
    break
```
