```python
# default args
def greet(name="ram"):
  print(name)
greet()
greet("chiku")
```

```python
# positional args
def greet(name,age):
  print(name," ",age)
greet("chiku",20)
```

```python
# keyword args
def student(name="Ram",roll=21): #default keywords Ram,21
  print(f"my name is {name} and my roll no is {roll}")
student("chiku",658) #positional argument
student(650,"ayush")
student(roll=674,name="chiku") #keyword arguments
student()
```

# **Scope of Variable**

```python
X=100 #global
def display():
  # global x
  x=200 #local
  print(x)

  def show():
    x=300
    print(x)
  show()
  print(x)

display()
print(x)
```

```python
x=100
def display():
  x=200
  print(x)
  def show():
    nonlocal x
    x=300

  show()
  print(x)
  print(x)
display()
print(x)
```

```python
x="global"
def outer():
  x="outer"
  def inner():
    x="inner"
    print(x)
  inner()
  print(x)
outer(x)
print(x)
```
