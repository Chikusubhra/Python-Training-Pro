```python
class Bank:
  def pay(self):
    pass

class UPI(Bank):
  def pay(self):
   print("payment is procceed by UPI payment method")
class phonePay(Bank):
  def pay(self):
   print("payment is proceed by phone-pay payment menthod")
class googlePay(Bank):
  def pay(self):
   print("payment is proceed by google-pay payment method")
c=[UPI(),phonePay(),googlePay()]

for i in c:
  i.pay()
```

```python
class vehicle:
  def start(self):
    pass
class car(vehicle):
  def start(self):
    print("car is start by the car key")
class bike(vehicle):
  def start(self):
    print("bike is stary by the bike key")
class scooty(vehicle):
  def start(self):
    print("scooty is start by the scooty key")
stat=[car(),bike(),scooty()]
for i in stat:
  i.start()
```

```python
class parent:
  def show(self):
    print("parent details")
class child(parent):
  def show(self):
    super().show()
    print("child details")
class child2(parent):
  def show(self):
    super().show()
    print("child2 details")
obj=child()
obj.show()

print("")
obj2=child2()
obj2.show()
```

```python
class Animal:
  def sound(self):
    pass
class dog(Animal):
  def sound(self):
    super().sound()
    print("I'm dog. barking ...")
class cat(Animal):
  def sound(self):
    super().sound()
    print("\nI'm cat. meow ...")
class peacock(Animal):
  def sound(self):
    super().sound()
    print("\nI'm peacock. mae-sow ...")

sounds=[dog(),cat(),peacock()]
for i in sounds:
  i.sound()
```

# ***doct-typing***

```python
class dog:
  def sound(self):
    print("dog sund")
class cat:
  def sound(self):
    print("cat meow")

def show(animal):
  animal.sound()

show(dog())
show(cat())
```

```python
# concept of docttyping concept..

class veg_food:
  def food(self):
    print("here only veg food is available ...")
class non_veg_food:
  def food(self):
    print("\nhere only non veg food is available ...")
class sea_food:
  def food(self):
    print("\nhere only sea food is avaiable ...")
def resturant(foods):
  foods.food()

resturant(veg_food())
resturant(non_veg_food())
resturant(sea_food())
```

```python
# concept of docttyping concept..

class SBI_Bank:
  def loan(self):
    print("here only 10% loan is available ...")
class AXIS_Bank:
  def loan(self):
    print("\nhere only 15% loan is available ...")
class BOI_Bank:
  def loan(self):
    print("\nhere only 20% loan is avaiable ...")
def BANK(banks):
  banks.loan()

BANK(SBI_Bank())
BANK(AXIS_Bank())
BANK(BOI_Bank())
```

# ***abstraction***

```python
from abc import ABC,abstractmethod

class vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    def stop(self):
        print("stopping...")

class bike(vehicle):
    def start(self):
        print("Bike starts by kick")

class car(vehicle):
    def start(self):
        print("car starting by key")
b=bike()
b.start()
b.stop()

c=car()
c.start()
c.stop()
```

```python
class bank:
  @abstractmethod
  def payment(self):
    pass
  def receipt(self):
    print("this is here your receipt...")
class sbi(bank):
  def payment(self):
    print("here is sbi account")
class axis(bank):
  def payment(self):
    print("here is axis account")
class boi(bank):
  def payment(self):
    print("here is boi account")
s=sbi()
s.payment()
s.receipt()
print("")
a=axis()
a.payment()
a.receipt()
print("")
b=boi()
b.payment()
b.receipt()
```
