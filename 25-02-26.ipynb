# **inheritance**             
*   single
*   multilevel
*   multiple
*   hierarchical
*   hybrid



```python
# @title
# single inheritance
class parent:
  def f1():
    print("i'm in parent class")
class child(parent):
  def d1():
    print("i'm in child class")
c=child
c.f1()
c.d1()
```

```python
# @title
# multilevel
class parent:
  def f1():
    print("i'm in parent class")
class child1(parent):
  def d1():
    print("i'm in child1 class")
class child2(child1):
  def d2():
    print("i'm in child2 class")
c=child2
c.f1()
c.d1()
c.d2()
```

```python
# @title
# multiple
class parent1:
  def f1():
    print("i,m in parent1 class")
class parent2:
  def f2():
    print("i'm in parent2 class")
class child(parent1,parent2):
  def d1():
    print("i'm in child class")
c=child
c.f1()
c.f2()
c.d1()
```

```python
# @title
# hierarchical
class parent:
  def f1():
    print("i'm in parent class")
class child1(parent):
  def d1():
    print("i'm in child1 class")
class child2(parent):
  def d2():
    print("i'm in child2 class")
class child3(parent):
  def d3():
    print("i'm in child3 class")
c1=child3
c1.d3()
c2=child2
c2.d2()
c3=child1
c3.d1()
```

```python
# @title
class parent:
  def __init__(self):
    print("parent")
  def f1(self):
    print("hi function 1")
class child(parent):
  def __init__(self):
    super().__init__()
    print("child")
c=child()
c.f1()
```

```python
# @title
# multiple
class mother:
  def __init__(self):
    print("mother knows cooking")
    super().__init__()
class father:
  def __init__(self):
    print("father knows cooking")
    super().__init__()
class child(mother,father):
  def __init__(self):
    print("child knows both")
    super().__init__()
c=child()
```

```python
class Student:
    def students(self):
        self.name = input("Enter your name: ")
        self.sec = input("Enter your section: ")
        self.roll = input("Enter your roll number: ")


class Academic(Student):
    def academic_marks(self):
        print("\nEnter your marks out of 30")
        self.java = int(input("Enter the Java mark: "))
        self.python = int(input("Enter the Python mark: "))
        self.c = int(input("Enter the C mark: "))
        self.os = int(input("Enter the mark for OS: "))

        total = self.java + self.python + self.c + self.os

        if total >= 90:
            self.academic_grade = "A"
        elif total >= 80:
            self.academic_grade = "B"
        elif total >= 70:
            self.academic_grade = "C"
        elif total >= 60:
            self.academic_grade = "D"
        else:
            self.academic_grade = "F"


class Sports(Student):
    def sports_marks(self):
        self.sm = int(input("\nEnter your sports score: "))

        if self.sm >= 90:
            self.sports_grade = "A"
        elif self.sm >= 80:
            self.sports_grade = "B"
        elif self.sm >= 70:
            self.sports_grade = "C"
        elif self.sm >= 60:
            self.sports_grade = "D"
        else:
            self.sports_grade = "F"


class Result(Academic, Sports):
    def finalreport(self):
        self.students()
        self.academic_marks()
        self.sports_marks()

        print("\n----- Final Report -----")
        print("Name:", self.name)
        print("Section:", self.sec)
        print("Roll No:", self.roll)
        print("Java:", self.java)
        print("Python:", self.python)
        print("C:", self.c)
        print("OS:", self.os)
        print("Sports Score:", self.sm)
        print("Academic Grade:", self.academic_grade)
        print("Sports Grade:", self.sports_grade)


r = Result()
r.finalreport()
```
