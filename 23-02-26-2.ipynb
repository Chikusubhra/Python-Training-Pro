```python
class Student:
    def students(self):
        self.name = input("Enter your name: ")
        self.sec = input("Enter your section: ")
        self.roll = input("Enter your roll number: ")
        print("\nStudent Details Collected")

class Academic(Student):
    def academic_marks(self):
        self.mark = int(input("Enter your academic marks: "))
        if self.mark >= 90:
            self.academic_grade = "A"
        elif self.mark >= 80:
            self.academic_grade = "B"
        elif self.mark >= 70:
            self.academic_grade = "C"
        elif self.mark >= 60:
            self.academic_grade = "D"
        else:
            self.academic_grade = "F"

class Sports(Student):
    def sports_marks(self):
        self.sm = int(input("Enter your sports score: "))
        if self.sm >= 90:
            self.sports_grade = "A"
        elif self.sm >= 80:
            self.sports_grade = "B"
        elif self.sm >= 70:
            self.sports_grade = "C"
        elif self.sm >= 60:
            self.sports_grade = "D"
        else:
            self.sports_grade = "F"

class Result(Academic, Sports):
    def finalreport(self):
        self.students()
        self.academic_marks()
        self.sports_marks()

        print("\n----- Final Report -----")
        print("Name:", self.name)
        print("Section:", self.sec)
        print("Roll No:", self.roll)
        print("Academic Grade:", self.academic_grade)
        print("Sports Grade:", self.sports_grade)
        print("Final report generated successfully!")


r = Result()
r.finalreport()

```
