```python
stu={
    "name":"ram",
    "roll":23,
    "section":"j"
}

x=stu["name"]
print(x)
x=stu.get("section")
print(x)
x=stu.values()
print(x)
stu["section"]="h"
print(stu)
stu.update({"name":"ayush"})
print(stu)
stu.update({"college":"giet"})
print(stu)
```

```python
stu={
     1:'a',
    2:'b',
    3:'c',
    "lists":[1,2,'chiku'],
     "name":"ram",
    "roll":23,
    "section":"j"
}

x=stu["name"]
print(x)
x=stu.get("section")
print(x)
x=stu.values()
print(x)
stu["section"]="h"
print(stu)
stu.update({"name":"ayush"})
print(stu)
stu.update({"college":"giet"})
print(stu)
del stu["lists"]
print(stu)
stu.clear()
print(stu)
del stu
print(stu)
```

```python
book={
    'name':"journey of love",
    'author':"Richard",
    "content":"Love",
    "date of publication":2014
}
print(book)
for key,value in book.items():
  print(key,":",value)
```

```python
shops={
    "orange":30,
    "banana":60,
    "apple":150
}
empty=0
for key,value in shops.items():
  print(value)
  empty+=value

print("total: ",empty)
```
