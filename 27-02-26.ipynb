# **polymorphism**

```python
"""
poly->many
morphism->forms
polymorphism is 2 types(runtime,complile times)
"""
```

```python
class Bank:
  def details(self):
    print("sbi")

class current(Bank):
  def intrest(self):
   print("current 30%")
class saving(Bank):
  def intrest(self):
   print("saving 20%")
class fixed(Bank):
  def intrest(self):
   print("fixed 10%")
c=[current(),saving(),fixed()]

for i in c:
  i.intrest()
```
