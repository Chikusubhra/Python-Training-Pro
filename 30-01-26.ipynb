```python
choice=int(input("enter you choice: \n1.tea\n2.coffee\n3.snacks\n"))
match choice:
  case 1:
    print("your tea is ready")
  case 2:
    print("your coffee is ready")
  case 3:
    print("your snacks is ready")
  case __:
    print("sorry, invalid input ...")
```

```python
choice=input("enter you favourite day: \n1.sun\n2.mon\n3.tue\n4.wed\n5.thr\n6.fri\n7.sat\n")
match choice:
  case "sun":
    print("Sunday is a sleepy day")
  case "mon":
    print("Monday is a boring day")
  case "tue":
    print("Tuesday is a charming day")
  case "wed":
    print("Wednesday is a chilling day")
  case "thr":
    print("Thursday is a good day")
  case "fri":
    print("Friday is a boring day")
  case "sat":
    print("Saturday is a happy day")
  case __:
    print("you should input which is shown")
```

```python
signal=input("enter traffic light: ")
match signal:
  case "❤️" | "red":
    print("stop")
  case "💛" | "🧡" | "yellow" | "orange":
    print("get ready step go")
  case "💚" | "green":
    print("you can go now")
  case __:
    print("traffic exception")
```

```python
character=input("enter a character: ").lower()
if len(character) !=1:
  print("must be a single letter")
else:
  match character:
    case "a" | "e" | "i" | "o" | "u":
      print("it is vowel")
    case __:
      print("it is consonant")
```

```python
num = int(input("enter number: "))
match num:
  case 0:
    print("Zero")
  case _ if num > 0:
    print("positive")
  case __:
    print("negative")
```

```python
num=int(input("enter number: "))
match num % 2:
  case 0:
    print("even")
  case __:
    print("odd")
```
